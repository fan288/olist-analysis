```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
import re

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# combine all the comment into one string
all_reviews = ' '.join(order_comment_dist_nona['review_comment_message'].tolist())

# use re
words = re.findall(r'\b\w+\b', all_reviews)

# delete the stop words
filtered_words = [word for word in words if word.lower() not in stop_words]

# count the frequence
word_counts = Counter(filtered_words)

# output the top 20 results
top_words = word_counts.most_common(20)
print(top_words)

# the result:
 #[('product', 15871), ('delivery', 5316), ('arrived', 4937), ('time', 4845),
 #('good', 4452), ('recommend', 4422), ('received', 4403), ('delivered', 4268),
 #('deadline', 3075), ('came', 2824), ('bought', 2770), ('quality', 2510), ('well', 2385),
 #('store', 2259), ('Product', 2195), ('one', 2112), ('fast', 1901), ('purchase', 1849), ('receive', 1610), ('great', 1575)]
```
